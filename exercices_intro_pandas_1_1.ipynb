```python
# nous avons un dictionnaire de clé-valeurs

dicos = {"Kaki": [20,25,23], "Kiwi":[34,56,7], "coconut":[12,9,25]}
```

```python
# affichons

print(dicos)
```

```python
# attribuons à chaque ligne de notre dico, un mois pour les valeurs
fruit_vendus = pd.DataFrame(dicos, index= ["Octobre sales", "Novembre sales", "decembre sales"])
display(fruit_vendus)
```

```python
# trouvons la somme des fruits du mois de octobre
somme_october = fruit_vendus.iloc[:1].sum(axis=1)
print(somme_october)
```

```python
# somme des fruits de chaque mois

somme = fruit_vendus.sum(axis=1)
print(somme)
```

```python
# ressortons les valeurs pour la colonne kaki
col_kaki = "Kaki"
print(fruit_vendus[col_kaki].values)
```

```python
#afficher la liste des colonnes du dataframe
print(fruit_vendus.columns)
```

```python
#afficher pour chaque colonne les valeurs
col_fruits= fruit_vendus.columns
for col in col_fruits:
    fruit_val = fruit_vendus[col].values
    print(col, fruit_val)
```
