# Mission 4

Define a list, then ask the user for a string. Then, insert the string in the list at a random position, and display the resulting list.

For example, with the list `test_list4 = ["p", "y", "t", "h", "o"]` and the input `n`, you *might* get:
```
["p", "n", "y", "t", "h", "o"]
```
... but you might also get:
```
["p", "y", "t", "n", "h", "o"]
```


````python
import random # pour avoir les element à des positions aléatoires
````

````python
test_list4 = ["p", "y", "t", "h", "o"]
# l'utilisateur choisit un string
user_string = input("insert the string: ")
#générer une position random
last_position = len(test_list4) - 1
# choisir une position aléatoire entre la 1ere et la dernière
position = random.randint(0,last_position)

# inserer la position et le string de l'utilisateur

test_list4.insert(position, user_string)
# la liste résultante

print(f"liste résultante est : {test_list4}")
````

# Mission 7

Define three lists containing no duplicates. Then, indicate (however you like) the two that have the most elements in common (or all three if there's a tie).

For instance, with:
```
list_a = [1, 2, 4, 8, 16, 32]
list_b = [1, 2, 3, 5, 8, 13]
list_c = [2, 3, 5, 7, 11, 13]
```
It may display:
```
Lists B and C have the most common elements!
```

````python
list_a = [1, 2, 4, 8, 16, 32]
list_b = [1, 2, 3, 5, 8, 13]
list_c = [2, 3, 5, 7, 11, 13]
# nous allons calculer le nbre d'element en commun entre list_a et list_b
commun_ab = 0
for value in list_a:
    if value in list_b:
        commun_ab +=1
print(f" On a {commun_ab} elements commun entre list_a et list_b")

# nous allons calculer le nbre d'element en commun entre list_a et list_c
commun_ac = 0
for value in list_a:
    if value in list_c:
        commun_ac +=1
print(f" On a {commun_ac} elements commun entre list_a et list_c")

# nous allons calculer le nbre d'element en commun entre list_b et list_c
commun_bc = 0
for value in list_b:
    if value in list_c:
        commun_bc +=1
print(f" On a {commun_bc} elements commun entre list_b et list_c")

# trouver la combinaison qui a le maximum d'elements

max_commun = max(commun_bc, commun_ab, commun_ac)

print(f"La combinaison qui a le maximum d'elements est : {max_commun}")

#condition pour afficher les deux list ayant le plus d'element en commun
if max_commun == commun_bc:
    print(f"Les listes B et C ont le plus d'elements commun !")
elif max_commun == commun_ab:
    print(f"Les listes A et B ont le plus d'elements commun !")
else:
    print(f"Les listes A et C ont le plus d'elements commun !")
````
